```python
from ultralytics import YOLO
import torch

# ตรวจสอบ GPU
print("CUDA:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# โหลดโมเดล
model = YOLO("yolo11n.pt")

# เทรน
results = model.train(
    data="data.yaml",

    # Training
    epochs=100,
    imgsz=640,
    batch=16,

    # GPU
    device=0,

    # DataLoader
    workers=8,

    # Output
    project="models",
    name="plate_yolo11n_v1",
    exist_ok=True,

    # Save
    save=True,
    save_period=10,

    # Optimization
    optimizer="auto",
    lr0=0.01,

    # Augmentation
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,

    degrees=5,
    translate=0.1,
    scale=0.3,

    fliplr=0.5,
    mosaic=1.0,
    mixup=0.0,

    # Validation
    val=True,
    plots=True,

    # Reproducibility
    seed=42
)
```
